```python
#import nympy 
import numpy

a = [7, 3, 5, 1, 9, 2, 6, 4, 8, 0]

#numpy has built in functions for mean, median, standard deviation, variance  and many more
mean = numpy.mean(a) #mean is the average of the list
median = numpy.median(a) #median is the middle value of the sorted list 
std = numpy.std(a) #standard deviation is the square root of the variance 
var = numpy.var(a) #variance = std^2 is the average of the squared differences from the mean 

print("mean: ", mean) #numpy has a built in mean function
print("median: ", median) #numpy has a built in median function
print("standard deviation: ", std) #numpy has a built in standard deviation function
print("variance: ", var) #numpy has a built in variance function
```
